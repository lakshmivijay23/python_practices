O(n) time complexity — only one loop.

O(1) space complexity — uses just two variables.

```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        min_price=float('inf')
        max_profit=0
        for price in prices:
            if price < min_price:
                min_price=price
            else:
                max_profit=max(max_profit, price-min_price)
        return max_profit
```

```python
obj=Solution()
max_profit=obj.maxProfit([3,5,8,1,9])
max_profit
```
