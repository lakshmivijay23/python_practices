**check prime**

```python
import math

num = 29
if num > 1:
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number")
            break
    else:
        print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

***function decorator***

```python
def welcome(fx):
  def mfx(*t, **d):
    print("Before hello function")
    fx(*t, **d)
    print("After hello function")
  return mfx

@welcome
def hello():
  print("Hello")

@welcome
def sums(a, c):
  print(a+c)
```

```python
 sums(a=5, c=6)
```

Write a decorator called @print_before_after that prints:

Before calling the function
After calling the function


every time any function is called.

```python
def print_before_after(fx):
  def wrapper():
    print("Before calling the function")
    fx()
    print("After calling the function")
  return wrapper

@print_before_after
def hi():
  print("Hi!")
```

```python
hi()
```

Create a decorator @timer that measures and prints how long a function takes to execute.

```python
import time

def timer(fx):

  def wrapper():
    start=time.time()
    fx()
    total_time=time.time()-start
    print(f"{fx.__name__} took {total_time:.2f} seconds")
  return wrapper

@timer
def slow_function():
  time.sleep(2)

```

```python
slow_function()
```

Write a decorator @log_function that prints the name of the function being executed.

```python
def log_function(fx):
  def wrapper():
    print(f"Executing {fx.__name__}")
    fx()
    print(f"Finished executing {fx.__name__}")
  return wrapper

@log_function
def printing():
   print("Printed")
```

```python
printing()
```

Write a decorator @validate_positive that checks if all numeric arguments passed to the function are positive.
If any are negative, print "Invalid argument" and donâ€™t run the function

```python
def validate_positive(fx):
  def wrapper(*args, **kwargs):
    for i in args:
      if i<0:
        print("Invalid argument")
    return fx(*args, **kwargs)
  return wrapper


@validate_positive
def check(a, c):
  return "Finish!"


# """
# Note:

# *args collects all positional arguments into a tuple
# **kwargs collects all keyword arguments into a dictionary

# """
```

**cache decorator**

```python
def cache(fx):
    memory_box = {}

    def wrapper(*args):
        print(*args)
        if args in memory_box:
            print(f"ðŸ§  I already know the answer for {args}!")
            return memory_box[args]

        print(f"ðŸ”¢ Calculating new answer for {args}")
        result = fx(*args)
        memory_box[args] = result
        return result
    return wrapper
```

```python
@cache
def fib(n):
    if n < 2:
        return n
    return fib(n - 1) + fib(n - 2)
```

```python
print(fib(2))
```

**class method decorator**

```python
def log_method(func):
  def wrapper(self, *args, **kwargs):
    print(f"Calling {func.__name__}")
    return func(self, *args, **kwargs)
  return wrapper

class Test:
  @log_method
  def hello(self):
    print("Hello from class!")
```

**parameterized access control**

```python
current_user = {"role":"guest"}

def role_required(role):
  def decorator(fx):
    def wrapper(*args, **kwargs):
      if current_user["role"]==role:
        return fx(*args, **kwargs)
      else:
        print(f"Access Denied: Requires {role} role")
    return wrapper
  return decorator

@role_required("admin")
def delete_user():
  print("User deleted!")
```

```python
delete_user()
```

**Class decorator**

```python
class Calculator:
  def __init__(self, func):
    self.function=func
  def __call__(self, *t, **d):
    result=self.function(*t, **d)
    return result**2

@Calculator
def add(a, b):
  return a-b

add(10, 2)
```
