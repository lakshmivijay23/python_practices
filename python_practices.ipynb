```python
import decimal

integer=5
print("Type of input:\n")
print(type(integer))
print("Type after conversion:\n")
print(decimal.Decimal(integer))
print(type(decimal.Decimal(integer)))
```

```python
vowel=['a', 'e', 'i', 'o', 'u']
string="programming"
count=0
for i in vowel:
  if i in string:
    count+=1
print(count)
```

```python
n=5
for i in range(n):
  for j in range(i, 5):
    print(" ", end=" ")
  for j in range(2*i+1):
    print("*", end=" ")
  print()
```
