**adding two list elements together**

```python
list1=[2,4,5]
list2=[7,2,1]
sum_list=[]
# check if the len of two list is the same
check = True if len(list1)==len(list2) else False
if check:
  for i in range(len(list1)):
    sum_list.append(list1[i]+list2[i])
```

```python
sum_list
```

```python
#simplify this using a list comprehension and zip()

new_list=[i+j for i, j in zip(list1, list2)]
new_list
```

**checking for palindrome using extended slicing technique**

```python
str1='Kayak'.lower()
if str1==str1[::-1]:
  print(True)
else:
  print(False)
```

```python
str1='Kayak'
str1= [i for i in str1.lower()]
flag=False
for i in range(len(str1)//2):
  if str1[i]==str1[-(i+1)]:
    flag=True
  else:
    flag=False
    break

if flag==True:
  print("Is palindrome")
else:
  print('Not a palindrome')
```

**counting digits, letters, and white spaces in a string**

```python
import re

string="Hello Im 123"

alpha = len(re.sub('[^A-Za-z]', "", string))
digits = len(re.sub('[^0-9]', '', string))
spaces = len(re.findall(r'[ /s]', string))

print(alpha)
print(digits)
print(spaces)
```
